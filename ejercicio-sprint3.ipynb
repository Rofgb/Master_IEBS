```python
# SPRINT 3

# Este ejercicio esta compuesto de lo siguiente:
#
# - funciones auxiliares para tratar las metricas
# - programa principal, con:
#     # definicion de constantes 
#     # llamada a las funciones auxiliares
#     # impresion por pantalla de los valores de las metricas
#
#
import csv
import json
import sys

# Funcion que llama al resto de funciones. 
# Se procesa el CSV una sola vez para que no tenga que recorrerlo cada
# vez que se pide un dato. Usaremos distintas variables para almacenar
# los datos segun lo que se pide en cada metrica.
def obtener_valores_sprint3(csv_file, metricas):
    for row in csv_file:
        for metrica in metricas:
            actualiza_lista_valores(row, metrica)
            actualiza_maximo(row, metrica)
            actualiza_minimo(row, metrica)
    
# Funcion auxiliar para obtener el valor de apertura de una fila
def actualiza_lista_valores(row, metrica):
    valor_fila = row.get(metrica.get(label_nombre))
    valor_actual = metrica.get(label_lista_valores)
    
    if valor_fila:
        valor_actual.append(float(valor_fila))
        
# Funcion auxiliar para comprobar maximos
def actualiza_maximo(row, metrica):
    dia_fila = row.get(label_fecha)
    valor_fila = float(row.get(metrica.get(label_nombre)))
    maximo_actual = metrica.get(label_maximo)
                    
    if valor_fila and dia_fila and maximo_actual.get(label_maximo) < valor_fila:
        maximo_actual.update({label_fecha: dia_fila, label_maximo: valor_fila})

# Funcion auxiliar para comprobar minimos
def actualiza_minimo(row, metrica):
    dia_fila = row.get(label_fecha)
    valor_fila = float(row.get(metrica.get(label_nombre)))
    minimo_actual = metrica.get(label_minimo)
                    
    if valor_fila and dia_fila and minimo_actual.get(label_minimo) > valor_fila:
        minimo_actual.update({label_fecha: dia_fila, label_minimo: valor_fila})

# Funcion auxiliar para crear las metricas asociadas a un dato, basadas en una misma plantilla
def crear_diccionario_metrica(nombre_metrica):
    return {label_nombre: nombre_metrica, \
            label_lista_valores: [], \
            label_maximo: {label_fecha: "", label_maximo: sys.float_info.min}, \
            label_minimo: {label_fecha: "", label_minimo: sys.float_info.max} \
           }

# PROGRAMA PRINCIPAL

label_fecha = "Date"
label_nombre = "Nombre"
label_lista_valores = "Lista_valores"
label_maximo = "Maximo"
label_minimo = "Minimo"

with open('TEF.csv', 'r') as file:
    csv_file = csv.DictReader(file)

    # Diccionario de metricas. Se usa la misma plantilla de diccionario
    # para todas las metricas
    metricas = [crear_diccionario_metrica("Open"), \
                crear_diccionario_metrica("Close"), \
                crear_diccionario_metrica("Volume")]
    
    obtener_valores_sprint3(csv_file, metricas)

    for metrica in metricas:
        print("\n\nValores de " + metrica.get(label_nombre) + ":\n")
        print(metrica.get(label_lista_valores))

        print("\n\nDia y valor en el cual el valor es maximo:")
        print(metrica.get(label_maximo))
        
        print("\n\nDia y valor en el cual el valor es minimo:")
        print(metrica.get(label_minimo))
    
    
# FIN PROGRAMA PRINCIPAL
```
