```python
def contains_taxi(taxi_id, dict_taxis):
    return any(key == taxi_id for key in dict_taxis.keys())
```

```python
#EJERCICIO 1

# [
#    {"taxi_id1" : {...}}, {"taxi_id2": {..., ..., ...} }, "taxi_id3": {} ...]


import csv
import json

def is_within_filter(row):
    return float(row.get("trip_total")) >= 20.0 \
           and float(row.get("trip_total")) <= 44.0 \
           and row.get("payment_type").upper() == "credit card".upper()

def is_valid_row(row):
    return row.get("unique_key") \
            and row.get("taxi_id") \
            and row.get("trip_total") \
            and row.get("payment_type")

with open('actividad-clase-2-actividad-clase-2.csv', 'r') as file:
    csv_file = csv.DictReader(file)
    dict_taxis = {}
    rows_list = []
    for row in csv_file:
        if not is_valid_row(row):
            print("fila con viaje " + row.get("unique_key") + " no es valida por lo que se descarta\n\n")
        else:    
            if is_within_filter(row) \
                and not contains_taxi(row.get("taxi_id"), dict_taxis):
                
                #empezar un nuevo diccionario para ese taxi
                dict_taxis.update({row.get("taxi_id") : row})

            if is_within_filter(row) \
                and contains_taxi(row.get("taxi_id"), dict_taxis):

                #anadir viaje al diccionario de viajes del taxi
                for key,value in dict_taxis.items():
                    if key == row.get("taxi_id"):
                        value.update(row)

    # Uso json que he buscado en internet para sacarlo bonito. Aun asi los Ids son muy largos asi que 
    # no se ve del todo bien en jupyter
    print("RESULTADO:")
    print(json.dumps(dict_taxis, indent=4))
```

```python
#EJERCICIO 2

import csv
import json

def is_within_filter(row):
    return (float(row.get("extras")) < 10.0 \
           or float(row.get("extras")) > 60.0) \
           and row.get("payment_type").upper() == "cash".upper()

def is_valid_row(row):
    return row.get("unique_key") \
            and row.get("taxi_id") \
            and row.get("trip_total") \
            and row.get("payment_type") \
            and row.get("extras")
    
with open('actividad-clase-2-actividad-clase-2.csv', 'r') as file:
    csv_file = csv.DictReader(file)
    dict_taxis = {}
    rows_list = []
    for row in csv_file:
        if is_valid_row(row):
            if is_within_filter(row) \
                and not contains_taxi(row.get("taxi_id"), dict_taxis):

                #empezar un nuevo diccionario para ese taxi
                dict_taxis.update({row.get("taxi_id") : row})

            if is_within_filter(row) \
                and contains_taxi(row.get("taxi_id"), dict_taxis):

                #anadir viaje al diccionario de viajes del taxi
                for key,value in dict_taxis.items():
                    if key == row.get("taxi_id"):
                        value.update(row)
        else:
            print("fila con viaje " + row.get("unique_key") + " no es valida por lo que se descarta\n\n")
            
    # Uso json que he buscado en internet para sacarlo bonito. Aun asi los Ids son muy largos asi que 
    # no se ve del todo bien en jupyter
    print("RESULTADO:")
    print(json.dumps(dict_taxis, indent=4))
```
